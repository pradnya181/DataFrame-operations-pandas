# Find out cars of USA based Origin 

```python
A[A.Origin == "USA"][["Make","Origin"]] # Cars of USA based Origin =48
```

# Find out Cars with no airbags and which are compact in type, from the output retrieve only Model, Mileage in city, Type, Airbags

```python
A[(A.AirBags == "None") & (A.Type == "Compact")][["Model","MPG.city","Type","AirBags"]]
```

# Que-Find out Models of cars having no AirBags and which are of compact type whose price is below 20.

```python
A[(A.AirBags == "None") & (A.Type == "Compact") & (A.Price < 20)][["Model"]]
```

# QUE-Find out cars with Mileage in city more than 15 which are compact in type.

```python
A[(A['MPG.city'] > 15) & (A.Type == "Compact")][['MPG.city',"Type","Model"]]
```

# Que-Find out Cars having mileage of more than 15 miles in city

```python
A[A['MPG.city'] > 15][["Model",'MPG.city']]
```

# QUE-Find out car having highest Mileage in city

```python
x = A['MPG.city'].max()
A[A['MPG.city'] == x][["Make"]]
```

# QUE-Find out top 7 most fuel economic cars for city, show results in sorted manner based on Horsepower.

```python
A.sort_values(by=["MPG.city"],ascending=True).tail(7)[["Model",'MPG.city',"Horsepower"]]
```

# QUE- Find out Models of cars which are having Price between 17 and 25-- > 27 cars

```python
A[(A.Price >= 17) & (A.Price <= 25)][["Model","Price"]]
```

# QUE-Find out top 7 most fuel economic cars for highway

```python
A.sort_values(by=["MPG.highway"]).tail(7)[["Manufacturer","MPG.highway"]]
```
